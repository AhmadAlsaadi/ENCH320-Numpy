# Introduction To Matplotlib 

Matplotlib is a powerful data visualization library for Python. It provides a wide range of tools and functions for creating various types of plots, charts, and graphs, making it an essential tool for data analysts, scientists, and researchers.

With Matplotlib, you can create line plots, scatter plots, bar plots, histograms, pie charts, and many other types of visualizations. It offers a high level of customization, allowing you to control every aspect of your plots, including colors, markers, line styles, labels, titles, and legends.

Matplotlib is designed to be easy to use and integrates well with other scientific libraries in the Python ecosystem, such as NumPy and Pandas. It provides a simple and intuitive interface, making it accessible to both beginners and experienced users.

One of the key features of Matplotlib is its ability to create publication-quality plots. You can save your plots in various file formats, including PNG, PDF, SVG, and more. This makes it easy to incorporate your visualizations into reports, presentations, and scientific publications.

Matplotlib is an open-source library and has a large and active community of users. This means that you can find plenty of resources, tutorials, and examples online to help you learn and master the library. Additionally, Matplotlib has extensive documentation that covers all its features in detail.

## Matplotlib Creator   

John Hunter was the creator of Matplotlib. He developed Matplotlib in 2002 as a tool for his neuroscience research. Hunter's goal was to create a library that would allow him and other researchers to create high-quality visualizations quickly and easily.

Under Hunter's leadership, Matplotlib grew in popularity and became an essential tool in the Python data science ecosystem. He actively maintained and contributed to the library until his untimely death in 2012. Hunter was highly regarded for his technical expertise, dedication to scientific visualization, and his support of the open-source community.

### Adding Title

We can add a title to the plot by using the function `plt.title()` where the title that we want to show on the plot is inserted as string to between the parenthesis of the function as it is shwon in the example below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.title("Graph of sin(x) function")
```

We can also change the marker size using the markersize parameter in the `plt.plot()` function as it is shwon in the example below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green",marker="o", markersize=15)
```

additionally, we can change the fill color of the marker using the parameter `markerfacecolor` inside the `plt.plot()` function as it is shown below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green",marker="o", markersize=15,markerfacecolor="y")
```

Observe that the markerfacecolor parameter changes only the fill which does not include the border color. To change the border color of the marker we can use the markeredgecolor parameter as it is shown in the example below: 

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green",marker="o", markersize=15,markerfacecolor="y", markeredgecolor="y")
```

if we want to change the width of the marker border we use the parameter `markeredgewidth` as it is shwon in the example below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green",marker="o", markersize=15,markerfacecolor="y", markeredgewidth=8)
```

5. **_Line Plot Transperancy_**   
Matplotlib allows you to control the transparency or opacity of lines in your plots. Adjusting line transparency can be useful for emphasizing certain elements, creating layered effects, or blending lines with the background. The line plot transperancy can be controlled through the parameter `alpha` which take a value from 0 to 1 where 0 means 100% transparent and 1 means 0% transperancy. The example below sets the transperancy level to 80%:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green",marker="o", markersize=15,markerfacecolor="y", alpha=.2)
```

### GridLines   
In order to add gridlines to ourline plot we use the function `plt.grid()` which by default enable the gridlines for both the x-axis and the y-axis as it is shown in the example below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green")
plt.grid()
```

We can customize the appearance of the gridlines by passing additional parameters to the `plt.grid()` function. Some commonly used parameters include:

* **_which_**: Specifies the gridlines to display. Options include 'major' (default), 'minor', or 'both'.
* **_axis_**: Specifies the axis for which to display the gridlines. Options include 'both' (default), 'x', or 'y'.
* **_color_**: Sets the color of the gridlines.
* **_linestyle_**: Specifies the style of the gridlines.
* **_linewidth_**: Sets the width of the gridlines.
* **_alpha_**: Controls the transparency of the gridlines.

Here's an example that customizes the gridlines:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green")
plt.grid(which='major', axis='both', color='k', linestyle='--', linewidth=1.5, alpha=0.2)
```
