# Introduction To Matplotlib 

Matplotlib is a powerful data visualization library for Python. It provides a wide range of tools and functions for creating various types of plots, charts, and graphs, making it an essential tool for data analysts, scientists, and researchers.

With Matplotlib, you can create line plots, scatter plots, bar plots, histograms, pie charts, and many other types of visualizations. It offers a high level of customization, allowing you to control every aspect of your plots, including colors, markers, line styles, labels, titles, and legends.

Matplotlib is designed to be easy to use and integrates well with other scientific libraries in the Python ecosystem, such as NumPy and Pandas. It provides a simple and intuitive interface, making it accessible to both beginners and experienced users.

One of the key features of Matplotlib is its ability to create publication-quality plots. You can save your plots in various file formats, including PNG, PDF, SVG, and more. This makes it easy to incorporate your visualizations into reports, presentations, and scientific publications.

Matplotlib is an open-source library and has a large and active community of users. This means that you can find plenty of resources, tutorials, and examples online to help you learn and master the library. Additionally, Matplotlib has extensive documentation that covers all its features in detail.

## Matplotlib Creator   

John Hunter was the creator of Matplotlib. He developed Matplotlib in 2002 as a tool for his neuroscience research. Hunter's goal was to create a library that would allow him and other researchers to create high-quality visualizations quickly and easily.

Under Hunter's leadership, Matplotlib grew in popularity and became an essential tool in the Python data science ecosystem. He actively maintained and contributed to the library until his untimely death in 2012. Hunter was highly regarded for his technical expertise, dedication to scientific visualization, and his support of the open-source community.

Some the missing information that the reader need to know includes, what is the x-axis and y-axis represent, what relation or topic does the plot trying to convery, etc. Therefore, matplotlib provide us with customization options that we can add to the plot, such as adding a title, labeling the axes, changing the line color or style, and adding markers to the data points etc. In this section we will go over most of the important options that matplotlib can add to plot.

### Adding Title

We can add a title to the plot by using the function `plt.title()` where the title that we want to show on the plot is inserted as string to between the parenthesis of the function as it is shwon in the example below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.title("Graph of sin(x) function")
```

### Adding x-axis and y-axis labels

We can add decriptions to the x-axis and y-axis through the use of `plt.xlabel()` and `plt.ylabel()` functions, respectively. The description should be inserted as string inside the functions parathesis as it is shown in the following example:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(20)
y=x**2
plt.plot(x,y)
plt.title("Graph of Velocity vs. Time")
plt.xlabel("Time (sec)")
plt.ylabel("Velocity (m/sec)")
```

2. **_Line style_**:     
we can change the line style in a plot by introducting the linestyle parameter in the `plt.plot()` as it it shown in the example below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green",linestyle="dashed")
```

In the example above, the word `dashed` in the linestyle parameter convert the plot line to dashed line. we could achieve the same result by replacing the wrod dashed with two dashes `--` as it is show in the example below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green",linestyle="--")
```

3. **_Data Point Marker_**   
   Matplotlib provides various marker styles that you can use to represent individual data points in your plots. These markers can be helpful in highlighting specific data points or adding visual cues to your plots. To customizes the data point marker, we introduce the parameter marker in `plt.plot()` function as we did with line color and line style. the example below shows a plot with filled circles as data point marker:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,color="green",marker="d")
```

4. **_Combined line color, linestyle and marker_**

We can combine the previous three customization options in a plot all at once by introducing them to `plt.plot()` function as it is shown in the example below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10)
y=np.sin(x)
plt.plot(x,y,"b--*")
```

In the example above, b represent the color of the line and the two dashes specify the linesyle as dashed line and the atstrisk represent the data point marker.
